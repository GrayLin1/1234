```
library(datasets)
```

```
library(data.table)
```

```
NBA1718$Position<-as.factor(NBA1718$Position)
NBA1718.mean<-data.table(NBA1718)[,.(TotalPointsMean=mean(TotalPoints,na.rm = T)),by=Position]
ggplot()+geom_bar(data = NBA1718.mean,aes(x=Position,y=TotalPointsMean),stat = "identity")
```

```
NBA1718.stat<-data.table(NBA1718)[,.(TotalPointsMean=mean(TotalPoints,na.rm = T),TotalPointsSD=sd(TotalPoints,na.rm = T)),by=Position]
ggplot(data = NBA1718.stat)+geom_bar(aes(x=Position,y=TotalPointsMean),stat = "identity")+geom_errorbar(aes(x=Position,ymin=TotalPointsMean-TotalPointsSD,ymax=TotalPointsMean+TotalPointsSD),width=.1)
```
